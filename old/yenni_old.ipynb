## Discrete-time competitive Beverton-Holt model

\begin{align}
N_{1,t+1}=\dfrac{r_1N_{1,t}}{1+a_{11}N_{1,t}+a_{12}N_{2,t}} \\
N_{2,t+1}=\dfrac{r_2N_{2,t}}{1+a_{22}N_{2,t}+a_{21}N_{1,t}}
\tag{1}
\end{align}

## Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from SALib.sample.morris import sample as ee_sample
from tqdm import tqdm
```

## Initialization

```python
# Variables for the time series
tmax = 100
time = np.linspace(1,tmax,tmax)
y1 = np.zeros(tmax)
#y1_t = np.zeros(tmax)
y2 = np.zeros(tmax)
#y2_t = np.zeros(tmax)
y1[0], y2[0] = 0.4, 0.6 # initial conditions
cov=np.zeros(0)

# Variables for the per capita growth rate
ymax = 100
y_increasing1 = np.linspace(1,ymax,ymax)
y_increasing2 = np.linspace(1,ymax,ymax)
per_capita1 = np.zeros(ymax)
per_capita2 = np.zeros(ymax)
y_posterior1 = np.zeros(ymax)
y_posterior2 = np.zeros(ymax)
```

## Test for only one set of parameters

```python
r1=15
r2=11
a11=0.7
a12=0.1
a21=0.1
a22=0.1
## Strength of stabilization
S1=r2/(1+(a12/a22)*(r2-1))
S2=r1/(1+(a21/a11)*(r1-1))                     
## Discrete-time competitive Beverton-Holt model
for i in range(tmax-1):
#    y1_t[i+1]=y1[i] # y1 in the previous step, y1[t-1]
#    y2_t[i+1]=y2[i] # y2 in the previous step, y2[t-1]
    y1[i+1] = r1 * y1[i] / ( 1 + a11 * y1[i] + a12 * y2[i] )
    y2[i+1] = r2 * y2[i] / ( 1 + a22 * y2[i] + a21 * y1[i] )

# Dummy plot
fig, ax = plt.subplots(figsize=(0.1, 0.1))
ax.plot(time, y1, alpha=0.5,linestyle="--", linewidth=10, markersize=12, color="r")
plt.rcParams.update({'font.size': 30})

fig, (ax1,ax2) = plt.subplots(2,figsize=(14, 10))
ax1.plot(time, y1, alpha=0.5,linestyle="--", linewidth=10, markersize=12, color="r", label='S1='+str(S1))
ax1.set(ylabel='$N_{1,t}$')
ax2.plot(time, y2, alpha=0.5,linestyle="--", linewidth=10, markersize=12, color="b", label='S2='+str(S2))
ax2.set(xlabel='Time (days)', ylabel='$N_{2,t}$')
plt.rcParams.update({'font.size': 30})
ax1.set_title('r1='+str(r1)+',r2='+str(r2)+',a11='+str(a11)+',a12='+str(a12)+',a21='+str(a21)+',a22='+str(a22), fontsize=25)
plt.tight_layout()
#plt.savefig('img/time_series1_r1_'+str(r1)+'_r2_'+str(r2)+'_a11_'+str(a11)+'_a12_'+str(a12)+'_a21_'+str(a21)+'_a22_'+str(a22)+'.png', format='png', dpi=50)
plt.show()
```

## Covariance: strength of stabilization and final population

```python
for r1 in range(15,21,5):
    for r2 in range(11,21,5):
        for a11 in np.arange(0.7,3.1,0.5):
            for a12 in np.arange(0.1,1.1,0.5):
                for a21 in np.arange(0.1,1.1,0.5):
                    for a22 in np.arange(0.1,1.1,0.5):
                        ## Strength of stabilization
                        S1=r2/(1+(a12/a22)*(r2-1))
                        S2=r1/(1+(a21/a11)*(r1-1))                     
                        ## Discrete-time competitive Beverton-Holt model
                        for i in range(tmax-1):
#                            y1_t[i+1]=y1[i] # y1 in the previous step, y1[t-1]
#                            y2_t[i+1]=y2[i] # y2 in the previous step, y2[t-1]
                            y1[i+1] = r1 * y1[i] / ( 1 + a11 * y1[i] + a12 * y2[i] )
                            y2[i+1] = r2 * y2[i] / ( 1 + a22 * y2[i] + a21 * y1[i] )
                        S=np.array([S1,S2])
                        abundance=np.array([y1[-1],y2[-1]])
                        cov=np.append(cov,np.cov(abundance, S)[0][1])
```

## Time series

```python
fig, (ax1,ax2) = plt.subplots(2,figsize=(14, 10))
ax1.plot(time, y1, alpha=0.5,linestyle="--", linewidth=10, markersize=12, color="r", label='S1='+str(S1))
ax1.set(ylabel='$N_{1,t}$')
ax2.plot(time, y2, alpha=0.5,linestyle="--", linewidth=10, markersize=12, color="b", label='S2='+str(S2))
ax2.set(xlabel='Time (days)', ylabel='$N_{2,t}$')
plt.rcParams.update({'font.size': 30})
ax1.set_title('covariance='+str(np.cov(abundance, S)[0][1])+'\n'+'r1='+str(r1)+',r2='+str(r2)+',a11='+str(a11)+',a12='+str(a12)+',a21='+str(a21)+',a22='+str(a22), fontsize=25)
plt.tight_layout()
plt.savefig('img/time_series_r1_'+str(r1)+'_r2_'+str(r2)+'_a11_'+str(a11)+'_a12_'+str(a12)+'_a21_'+str(a21)+'_a22_'+str(a22)+'.png', format='png', dpi=50)
plt.show()
```

## Increasing N1 and N2

```python
## Discrete-time competitive Beverton-Holt model
for i in range(ymax):
    per_capita1[i] = r1 / ( 1 + a11 * y_increasing1[i] + a12 * y_increasing2[i] )
    per_capita2[i] = r2 / ( 1 + a22 * y_increasing2[i] + a21 * y_increasing1[i] )
    y_posterior1[i] = y_increasing1[i] * per_capita1[i]
    y_posterior2[i] = y_increasing2[i] * per_capita2[i]
```

## Finite Growth Rate and Stock Recruitment Curve

```python
N = np.linspace(0,ymax,ymax)
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,figsize=(14, 10))
# Finite Growth Rate
ax1.plot(N, per_capita1, alpha=0.5,linestyle="dotted", linewidth=10, markersize=12, color="r")
ax1.axhline(y=1, color='grey', linestyle='-')
ax1.set(ylabel='$N_{1,t+1}/N_{1,t}$')
# Stock-Recruitment
ax2.plot(N, y_posterior1, alpha=0.5,linestyle="dotted", linewidth=10, markersize=12, color="y")
ax2.plot(N,N, color="grey") # add a straight line
ax2.set_ylim([-1, 10])
ax2.set(xlabel='$N_{1,t}$', ylabel='$N_{1,t+1}$')
# Finite Growth Rate
ax3.plot(N, per_capita2, alpha=0.5,linestyle="dotted", linewidth=10, markersize=12, color="b")
ax3.axhline(y=1, color='grey', linestyle='-')
ax3.set(ylabel='$N_{2,t+1}/N_{2,t}$')
# Stock-Recruitment
ax4.plot(N, y_posterior2, alpha=0.5,linestyle="dotted", linewidth=10, markersize=12, color="g")
ax4.plot(N,N, color="grey") # add a straight line
ax4.set_ylim([-1, 20])
ax4.set(xlabel='$N_{2,t}$', ylabel='$N_{2,t+1}$')
plt.rcParams.update({'font.size': 30})
ax1.set_title('covariance='+str(np.cov(abundance, S)[0][1])+'\n'+'r1='+str(r1)+',r2='+str(r2)+',a11='+str(a11)+',a12='+str(a12)+',a21='+str(a21)+',a22='+str(a22), fontsize=25)
plt.tight_layout(pad=0)
plt.savefig('img/finite_growth_rate_r1_'+str(r1)+'_r2_'+str(r2)+'_a11_'+str(a11)+'_a12_'+str(a12)+'_a21_'+str(a21)+'_a22_'+str(a22)+'.png', format='png', dpi=50, bbox_inches='tight')
plt.show()
```
