## Discrete-time competitive Beverton-Holt model

\begin{align}
N_{1,t+1}=\dfrac{r_1N_{1,t}}{1+a_{11}N_{1,t}+a_{12}N_{2,t}} \\
N_{2,t+1}=\dfrac{r_2N_{2,t}}{1+a_{22}N_{2,t}+a_{21}N_{1,t}}
\tag{1}
\end{align}

```python
# calculate log per capita growth rate
def getPCG(r1, r2, a11, a12, a21, a22, N):
    # r is a list of fecundities for species 1 and 2
    # alpha is a matrix of interaction coefficients where alpha[i][j] is the effect of j on i
    # N is a list of densities for species 1 and 2
    newN = np.zeros(2)
    newN[0] = r1 * N[0] / (1 + a11 * N[0] + a12 * N[1])
    newN[1] = r2 * N[1] / (1 + a21 * N[0] + a22 * N[1])
    out = np.zeros(2)
    out[0] = np.log(newN[0]) - np.log(N[0])  # per capita growth rate of species 1
    out[1] = np.log(newN[1]) - np.log(N[1])  # per capita growth rate of species 2
    return out
```

```python
# figure out the equilibrium density of the focal species' competitor
# given a fixed density of the focal species
def getEqDensity(fixed_species, r1, r2, a11, a12, a21, a22, N_star):
    # fixed_species = focal species (this one is fixed, density of the other is returned)
    # N_start (a scalar) is the density of the fixed species
    if fixed_species == 1: # density of species 1 is known
        N2_star = (r2 - 1 - a21*N_star) / a22
        return N2_star
    elif fixed_species == 2: # density of species 2 is known
        N1_star = (r1 - 1 - a12*N_star) / a11
        return N1_star
    else:
        print("Invalid species number. Only 1 or 2 is allowed.")
        return None
```

```python
@jit(nopython=True)
def time_simul_fixed(fixed_species, r1, r2, a11, a22, a12, a21, N_star):
    y01,y02=N_star,N_star
    tmax = 2
    y1 = np.zeros(tmax)
    y2 = np.zeros(tmax)
    per_cap1 = np.zeros(tmax)
    per_cap2 = np.zeros(tmax)
    y1_relat = np.zeros(tmax)
    y2_relat = np.zeros(tmax)
    y1[0], y2[0] = y01, y02 # initial conditions
    i, stopRun = 0, 0
    while stopRun == 0: # stops the simulation when the dynamics converges
        per_cap1[i] = r1 / (1 + a11 * y1[i] + a12 * y2[i])
        per_cap2[i] = r2 / (1 + a22 * y2[i] + a21 * y1[i])
        y1[i+1] = y1[i] * per_cap1[i]
        y2[i+1] = y2[i] * per_cap2[i]
        y1_relat[i] = y1[i] / (y1[i] + y2[i])
        y2_relat[i] = y2[i] / (y1[i] + y2[i])
        if fixed_species == 1:
            y1[i+1] = y01
            if np.abs(y2[i+1]-y2[i]) < 1.0e-11:
                stopRun = 1
        elif fixed_species == 2:
            y2[i+1] = y02
            if np.abs(y1[i+1]-y1[i]) < 1.0e-11:
                stopRun = 1
        else:
            raise ValueError("fixed_species should be either '1' or '2'")
        per_cap1 = np.append(per_cap1, per_cap1[i])
        per_cap2 = np.append(per_cap2, per_cap2[i])
        y1 = np.append(y1, y1[i])
        y2 = np.append(y2, y2[i])
        y1_relat = np.append(y1_relat, y1_relat[i])
        y2_relat = np.append(y2_relat, y2_relat[i])
        i += 1
        if i > 1000:
            break
    
    if fixed_species == 1:
        return y2[-1]
    elif fixed_species == 2:
        return y1[-1]
```

```python
def getNFD(r1, r2, a11, a12, a21, a22, lowN, deltaN):
    # lowN is the low density 
    # and deltaN is the increase in N for the higher density to explore

    # vectors for output
    pgr1, freq1, pgr2, freq2 = np.zeros(2), np.zeros(2), np.zeros(2), np.zeros(2)
    sp2_low1,sp1_low2,sp2_high1,sp1_high2=0.0,0.0,0.0,0.0

    # get low density growth rate for species 1
    tmpN = np.zeros(2)
    tmpN = np.zeros(2)
    tmpN[0] = lowN  # set density of focal species
    tmpN[1] = time_simul_fixed(fixed_species=1, r1=r1, r2=r2, a11=a11, a12=a12, a21=a21, a22=a22, N_star=tmpN[0])  # figure out density of its competitor
    tmpOut = getPCG(r1=r1, r2=r2, a11=a11, a12=a12, a21=a21, a22=a22, N=tmpN)
    pgr1[0] = tmpOut[0]
    freq1[0] = tmpN[0] / np.sum(tmpN)
    sp2_low1 = tmpN[1]

    # get low density growth rate for species 2
    tmpN = np.zeros(2)
    tmpN[1] = lowN
    tmpN[0] = time_simul_fixed(fixed_species=2, r1=r1, r2=r2, a11=a11, a12=a12, a21=a21, a22=a22, N_star=tmpN[1])
    tmpOut = getPCG(r1=r1, r2=r2, a11=a11, a12=a12, a21=a21, a22=a22, N=tmpN)
    pgr1[1] = tmpOut[1]
    freq1[1] = tmpN[1] / np.sum(tmpN)
    sp1_low2 = tmpN[0]

    # get higher density growth rate for species 1
    tmpN = np.zeros(2)
    tmpN[0] = lowN + deltaN
    tmpN[1] = time_simul_fixed(fixed_species=1, r1=r1, r2=r2, a11=a11, a12=a12, a21=a21, a22=a22, N_star=tmpN[0])
    tmpOut = getPCG(r1=r1, r2=r2, a11=a11, a12=a12, a21=a21, a22=a22, N=tmpN)
    pgr2[0] = tmpOut[0]
    freq2[0] = tmpN[0] / np.sum(tmpN)
    sp2_high1 = tmpN[1]

    # get higher density growth rate for species 2
    tmpN = np.zeros(2)
    tmpN[1] = lowN + deltaN
    tmpN[0] = time_simul_fixed(fixed_species=2, r1=r1, r2=r2, a11=a11, a12=a12, a21=a21, a22=a22, N_star=tmpN[1])
    tmpOut = getPCG(r1=r1, r2=r2, a11=a11, a12=a12, a21=a21, a22=a22, N=tmpN)
    pgr2[1] = tmpOut[1]
    freq2[1] = tmpN[1] / np.sum(tmpN)
    sp1_high2 = tmpN[0]

    out = {
        "pgr1": pgr1,
        "freq1": freq1,
        "pgr2": pgr2,
        "freq2": freq2
    }
    return out,sp2_low1,sp1_low2,sp2_high1,sp1_high2
```

```python
def coexist(y1,y2):
    coexist = y1>0.01 and y2>0.01
    if np.isnan(y1) or np.isnan(y2):
        coexist="nan"
    elif np.isinf(y1) or np.isinf(y2):
        coexist="inf"
    elif coexist==True:
        coexist = 1
    elif coexist==False:
        coexist = 0
    return coexist
```

```python
def start_above_zero(pgr1,pgr2):
    above_zero = pgr1 > 0 and pgr2 > 0
    if above_zero==True:
        above_zero = 1
    elif above_zero==False:
        above_zero = 0
    return above_zero
```

```python
def check_intercept(x1, y1, x2, y2):
    """Checks if two line segments intercept.

    Args:
        x1 (tuple): Tuple containing the x-coordinates of the endpoints of the first line segment.
        y1 (tuple): Tuple containing the y-coordinates of the endpoints of the first line segment.
        x2 (tuple): Tuple containing the x-coordinates of the endpoints of the second line segment.
        y2 (tuple): Tuple containing the y-coordinates of the endpoints of the second line segment.

    Returns:
        bool: 1 if the line segments intercept, 0 otherwise.
    """
    x_diff_1 = x1[1] - x1[0]
    y_diff_1 = y1[1] - y1[0]
    x_diff_2 = x2[1] - x2[0]
    y_diff_2 = y2[1] - y2[0]
    div = (y_diff_2 * x_diff_1) - (x_diff_2 * y_diff_1)

    if div == 0:
        return 0

    ua = ((x_diff_2 * (y1[0] - y2[0])) - (y_diff_2 * (x1[0] - x2[0]))) / div
    ub = ((x_diff_1 * (y1[0] - y2[0])) - (y_diff_1 * (x1[0] - x2[0]))) / div

    if 0 <= ua <= 1 and 0 <= ub <= 1:
        return 1

    return 0
```
